```python
import pandas as pd
```

```python
##Lê um arquivo ".csv", para ler um "xlsx" .read_excel()
df = pd.read_csv('football-events\ginf.csv')
dff = pd.read_csv('football-events\events.csv')
```

```python
#Decreve os dados do DataFrame
dff.head()
```

```python
##Contar valores de uma coluna do DatFrame
pd.value_counts(dff['side'])
```

```python
##Buscar dados com base em uma condição, método ".loc". Para mais condições, deve-se seguir o padrão e adicionar o operador &
dff.loc[(dff['event_team'] == 'Hamburg SV') & (dff['opponent'] == 'Borussia Dortmund')]
##Peguei os jogos do Hamburg SV contra o Borussia Dortmund
```

```python
dtf = dff.loc[(dff['event_team'] == 'Hamburg SV') & (dff['opponent'] == 'Borussia Dortmund')]
## .count() retorna a quantidade de registros contidas para cada coluna.
##Nesse caso, solicitei a quantidade de dados para os jogos que foram em casa
dtf[dtf['side'] == 1].count()
```

```python
##Plota um gráfico para uma coluna específica com o método .plot(). Quando não define os dois eixos, ele cria o eixo x baseado na quantidade de registros
df['odd_h'].plot()
```

```python
#Plotando um gráfico, definindo os dois eixos, o tipo do gráfico, o título e a cor.
df.plot(x='odd_h',y='fthg', kind='scatter', title='Odd_h x fthg',color='green');
#Para criar um histograma de frequencias, coloca dataframe['column'].hist(bins=100), define o tamanho do beans
```
